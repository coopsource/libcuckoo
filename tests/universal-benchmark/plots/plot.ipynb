```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.backends.backend_pdf import PdfPages

# Setup PDF output, one page for each graph (call to savefig)
pp = PdfPages('benchmarks.pdf')

tables = ('TBB (c_h_m)', 'libcuckoo (old API)', 'libcuckoo (new API)')
y_pos = np.arange(len(tables))
performance = (8, 10, 4)

plt.barh(y_pos, performance, align='center', alpha=0.4)
plt.yticks(y_pos, tables)

text = '''
Command Line = "universal-benchmark ..."
Machine Type = XYZ
Number of Threads = 2

KEY   = uint64_t
VALUE = uint64_t
'''

plt.title('libcuckoo vs. TBB performance comparison')
plt.xlabel('Throughput (Mops/sec)')

a = plt.text(11, 2.5, text, verticalalignment='top', family='monospace')

pp.savefig(bbox_extra_artists=(a,), bbox_inches='tight')
pp.close()

plt.show()
```
